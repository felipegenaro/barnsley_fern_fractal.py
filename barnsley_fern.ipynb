```python
# barnsley fern

# matrix notation
#
# f(x, y) = |a  b| |x|   |e|
#           |c  d| |y| + |f|

# -------------------------------

# transformations
#
# f1(x, y) = |0.00  0.00| |x| 
#            |0.00  0.16| |y| 

# f2(x, y) = | 0.85  0.04| |x|   |0.00|
#            |-0.04  0.85| |y| + |1.60|

# f3(x, y) = |0.20  -0.26| |x|   |0.00|
#            |0.23   0.22| |y| + |1.60|

# f4(x, y) = |-0.15  0.28| |x|   |0.00|
#            | 0.26  0.24| |y| + |0.44|
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# functions for the transformations

# every function there is a probability/proportion
# 01% - this portion will generate the stem
def function1(x, y):
    return (0., 0.16*y)

# 85% - this portion will generate the successively smaller leaflets
def function2(x, y):
    return (0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6)

# 07% - this portion will generate the largest left-hand leaflet
def function3(x, y):
    return (0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6)

# 07% - this portion will generate the largest right-hand leaflet
def function4(x, y):
    return (-0.15*x + 0.28*y, 0.26*x + 0.24*y + 0.44)

# list of all functions
functions = [function1, function2, function3, function4]
```

```python
points = 100000
x, y = 0, 0

x_list, y_list = [], []

# pick the function to transform the current point
for i in range(points):
    function = np.random.choice(functions, p=[0.01, 0.85, 0.07, 0.07])
    
    x, y = function(x, y)
    
    x_list.append(x)
    y_list.append(y)
```

```python
plt.scatter(x_list, y_list, s=0.15, color='green')
```
